## Univariate Analysis

### Subtask:
Analyze the distribution of individual columns. For numerical columns, use histograms and box plots. For categorical columns like 'Unit Name' and 'Year', use bar plots to visualize frequencies and counts.


**Reasoning**:
Having completed the univariate analysis for numerical crime-related columns, I will now create a bar plot for the 'Unit Name' column to visualize the frequency of occurrences for each unit, as per instruction 3 of the subtask.



```python
plt.figure(figsize=(12, 6))
sns.countplot(y=data['Unit Name'], order=data['Unit Name'].value_counts().index)
plt.title('Frequency of Crime Reports by Unit Name')
plt.xlabel('Count')
plt.ylabel('Unit Name')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Now, I will create a bar plot for the 'Year' column to visualize the frequency or count of occurrences for each year, as per instruction 4 of the subtask.



```python
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Year'], order=data['Year'].value_counts().index)
plt.title('Frequency of Crime Reports by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Reasoning**:
First, I need to group the data by 'Year' and sum all the crime-related columns to get yearly trends, as per the first instruction for time-series analysis.



```python
yearly_crime_trends = data.groupby('Year')[crime_columns + ['Total Cases']].sum().reset_index()
print("Yearly crime trends DataFrame:")
print(yearly_crime_trends.head())
```

**Reasoning**:
Now that the yearly crime trends have been aggregated, I will create a line plot to visualize the trend of 'Total Cases' over the years, as per instruction 2 of the subtask.



```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Total Cases', data=yearly_crime_trends)
plt.title('Total Crime Cases Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Cases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Reasoning**:
I will now iterate through the `crime_columns` list and create a line plot for each crime type, visualizing its trend over the years, as per instructions 3 and 4 of the subtask.



```python
for column in crime_columns:
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Year', y=column, data=yearly_crime_trends)
    plt.title(f'Trend of {column} Over the Years')
    plt.xlabel('Year')
    plt.ylabel(column)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## Time-series Analysis of Crime Trends

### Subtask:
Analyze how different crime types and 'Total Cases' have changed over the years by plotting time-series charts for various crime categories and 'Total Cases'.


**Reasoning**:
To perform geographical analysis, I will first group the data by 'Unit Name' and calculate the sum of 'Total Cases' for each unit, as this will help in identifying which units have the highest crime rates.



```python
unit_crime_summary = data.groupby('Unit Name')['Total Cases'].sum().sort_values(ascending=False).reset_index()
print("Top 10 Units by Total Cases:")
print(unit_crime_summary.head(10))
```

**Reasoning**:
Now that the total crime cases per unit have been calculated and sorted, I will create a bar plot to visualize the 'Total Cases' for each 'Unit Name'. This visualization will clearly show the crime distribution across different geographical units, addressing the next step in geographical analysis.



```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Total Cases', y='Unit Name', data=unit_crime_summary)
plt.title('Total Crime Cases by Unit Name')
plt.xlabel('Total Cases')
plt.ylabel('Unit Name')
plt.tight_layout()
plt.show()
```

**Reasoning**:
To begin the geographical analysis, I will group the data by 'Unit Name' and sum all the crime-related columns to get the total count for each crime type per unit, storing it in `unit_crime_types_summary`.



```python
unit_crime_types_summary = data.groupby('Unit Name')[crime_columns].sum().reset_index()
print("Crime type summary by Unit Name:")
print(unit_crime_types_summary.head())
```

**Reasoning**:
Now that `unit_crime_types_summary` has been created, I will iterate through each crime type in `crime_columns` and generate a bar plot. Each plot will show the total count for that crime type per 'Unit Name', sorted in descending order, with appropriate titles and labels. This addresses instructions 2, 3, 4 and 5 of the subtask.



```python
for column in crime_columns:
    plt.figure(figsize=(12, 8))
    # Sort by the current crime column in descending order
    sorted_data = unit_crime_types_summary.sort_values(by=column, ascending=False)
    sns.barplot(x=column, y='Unit Name', data=sorted_data)
    plt.title(f'Total {column} Cases by Unit Name')
    plt.xlabel(f'Total {column} Cases')
    plt.ylabel('Unit Name')
    plt.tight_layout()
    plt.show()
```

## Geographical/Unit-wise Analysis

### Subtask:
Compare crime statistics for specific crime types across different 'Unit Name' categories.


**Reasoning**:
The geographical analysis of crime types by unit has been completed. The next logical step, according to the main task, is to perform correlation analysis between the numerical crime types. I will start by computing the correlation matrix for the `crime_columns`.



```python
correlation_matrix = data[crime_columns].corr()
print("Correlation Matrix of Crime Types:")
print(correlation_matrix)
```

**Reasoning**:
Now that the correlation matrix has been computed, I will visualize it using a heatmap to graphically represent the correlations between different crime types. This will make it easier to identify strong positive or negative relationships, addressing the next logical step in correlation analysis.



```python
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Crime Types')
plt.tight_layout()
plt.show()
```

## Outlier Detection

### Subtask:
Identify and visualize outliers in the numerical crime columns using appropriate statistical methods or plots, such as box plots or Z-scores, to understand extreme values.

**Reasoning**:
Now that the Outlier Detection subtask has been introduced, I will calculate the Z-scores for each numerical crime column to identify outliers. Z-scores help quantify how far away a data point is from the mean in terms of standard deviations.



```python
from scipy.stats import zscore

for column in crime_columns:
    data[f'{column}_zscore'] = np.abs(zscore(data[column]))
    outliers = data[data[f'{column}_zscore'] > 3]

    print(f"\nOutliers in {column} (Z-score > 3):")
    if not outliers.empty:
        print(outliers[['Unit Name', 'Year', column, f'{column}_zscore']].head())
    else:
        print("No outliers found.")
```

**Reasoning**:
Now that Z-scores have been calculated and outliers identified, I will generate box plots for each numerical crime column to visually represent their distributions and further highlight the identified outliers, which directly addresses the visualization aspect of the subtask.



```python
for column in crime_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=data[column])
    plt.title(f'Box Plot of {column} with Outliers')
    plt.ylabel(column)
    plt.tight_layout()
    plt.show()
```
